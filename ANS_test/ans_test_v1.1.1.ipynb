```python
#import：
from IPython.display import display, Image, clear_output
import time

#send to google import:
import requests
from bs4 import BeautifulSoup
import json
import time

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok
    
#intro
print('Welcome to Approximation Number Sense(ANS) Test')
time.sleep(1.5)
print("In this test you will see one image with blue dots and yellow dots,")
time.sleep(1.5)
print("you have to choose which side has more dots.")
time.sleep(1.5)
print("REMEMBER, you need select: \na for left \nb for right")
time.sleep(3)
clear_output(wait=False)

print("Before taking the test, if you have any of the following please enter：\n a for coffee or tea or any energy drink; \n b for alcohol or any psychiatric drugs; \n c for none of them.")
situation = input('Have you drink any above?')
clear_output(wait=False)

print('test will start soon! \n Please notice that every test will take 3 second to response to each question \n if you exceed the time limit, you will not earn any points.\n The test will only take 3 minutes')
name = input('please enter your username to start')
clear_output(wait=False)

#store the images
image_1 = Image("Slide1.PNG", width = 600)
image_2 = Image("Slide2.PNG", width = 600)
image_3 = Image("Slide3.PNG", width = 600)
image_4 = Image("Slide4.PNG", width = 600)
image_5 = Image("Slide5.PNG", width = 600)
image_6 = Image("Slide6.PNG", width = 600)
image_7 = Image("Slide7.PNG", width = 600)
image_8 = Image("Slide8.PNG", width = 600)
image_9 = Image("Slide9.PNG", width = 600)
image_10 = Image("Slide10.PNG", width = 600)
image_11 = Image("Slide11.PNG", width = 600)
image_12 = Image("Slide12.PNG", width = 600)
image_13 = Image("Slide13.PNG", width = 600)
image_14 = Image("Slide14.PNG", width = 600)
image_15 = Image("Slide15.PNG", width = 600)
image_16 = Image("Slide15.PNG", width = 600)
image = [image_1, image_2, image_3, image_4, image_5, image_6, image_7,image_8, image_9,image_10,image_11,image_12,image_13,image_14,image_15,image_16]
ans = ['a','a','a','a','b','a','b','b','a','a','b','b','a','a','a','a']

# test
time_taken_list = []
score = 0
score_each = []
time_allstart = time.time()
for i in range(16):
    display(image[i])
    time.sleep(1.5)
    start_time = time.time()
    clear_output(wait=False)

    print("REMEMBER, you need select: \na for left \nb for right")
    ans_q = input("which side has more dots")
    
    end_time = time.time()
    time_taken = end_time - start_time
    time_taken_list.append(time_taken)
    if ans_q == ans[i] and time_taken <=3:
        print("Correct!")
        score = score + 1
        score_each.append(1)
    else:
        print("Sorry, you do not get it right in time")
        score_each.append(0)
    print("time_taken:", time_taken)
    total_score = score / 16 * 100
    print(f'score so far = {total_score}/100')
    time.sleep(1.5)
    clear_output(wait=False)
    
    expiration_time = time.time() - time_allstart
    if expiration_time >= 180:
        print("you have run out of the time.")
        break
        

#summary
print('Thank you for taking the test \nHere is your result')
print(f"total score = {total_score}")

average_time_taken = sum(time_taken_list)/16
print(f"average time taken: {average_time_taken}")
time.sleep(5)
clear_output(wait=False)
print(score_each)

#send to google form
print("Please read:")
print("")
print("we wish to record your response data")
print("to an anonymised public data repository. ")
print("Your data will be used for educational teaching purposes")
print("practising data analysis and visualisation.")
print("")
print("Please type yes in the box below if you consent to the upload.")
result = input("> ")
if result == "yes":
    print("Thanks - your data will be uploaded.")
    print('Could you please answer following questions:')
    age = input('What is your age?')
    sex = input('what is your sex?')
    results_dict = {
    'name': [],
    'situation':[],
    'score': [],
    'average_time_taken': [], 
    'age':[],
    'sex':[]
}
    results_dict['name'].append(name)
    results_dict['situation'].append(situation)
    results_dict['score'].append(total_score)
    results_dict['average_time_taken'].append(average_time_taken)
    results_dict['age'].append(age)
    results_dict['sex'].append(sex)

    form_url = 'https://docs.google.com/forms/d/e/1FAIpQLScvwiUN5D3HUxma_koghzRp_VFgk-H6gm0g83ikATnzDZLFVQ/viewform?usp=sf_link'
    send_to_google_form(results_dict, form_url)
#send_to_google_form(data_dict, form_url)
    
else:
    print("No problem we hope you enjoyed the test.")

import pandas as pd
results_dict = {
    'filename': ['Slide1', 'Slide2', 'Slide3', 'Slide4', 'Slide5', 'Slide6', 'Slide7', 'Slide8', 'Slide9', 'Slide10', 'Slide11', 'Slide12', 'Slide13', 'Slide14', 'Slide15', 'Slide16'],
    'nL': [12,14,18,20,9,9,12,7,13,9,7,14,9,10,20,18],
    'nR': [9,12,16,18,10,8,16,8,12,8,8,16,8,9,18,16],
    'ratio': ['4:3','7:6','9:8','10:9','9:10','9;8','3:4','7:8','13:12','9:8','7:8','7:8','9:8','10:9','10:9','9:8'],
    'correct': score_each}
mydataframe = pd.DataFrame(results_dict)
mydataframe 
mydataframe.to_csv('data.csv')
```
